```python
import os
import csv
import pydicom as dicom
```

```python
#first mount your drive 
# you can get file path by right click on it then choose copy path
# paste your folder path here to directory between ("                   ")   in the two lines below
  
from pydicom import dcmread
ds = dcmread("/content/drive/MyDrive/1_1.3.6.1.4.1.23849.2168902280.159.1637689365147349897")
File_name = os.fsdecode("/content/drive/MyDrive/1_1.3.6.1.4.1.23849.2168902280.159.1637689365147349897")
```

```python
Study = ds[0x0008,0x1030].value
Protocol = ds[0x0018,0x1030].value
Manufacturar = ds[0x0008,0x0070].value

try:
  ct_par1 = ds[0x0040,0xa730][12]     
  ct_par2 = ct_par1[0x0040,0xa730]
  ct_par3 = ct_par2[5]
  ct_par4 = ct_par3[0x0040,0xa730]  
  ct_par5 = ct_par4[0]
  ct_par5[0x0040,0xa300][0].NumericValue
except:
  pass

try:
  acq_protocol1 = ds[0x0040,0xa730][13]
  ct_acq_type2 = acq_protocol1[0x0040,0xa730]
except:
  pass

     
#observer 
try:
  observer_physical_location = ds[0x0040,0xa730][7]
  Device_Observar_physical_location = observer_physical_location.TextValue
except:
  Device_Observar_physical_location = "NOT available"



#exposure time 
try:
  expsure_time1 = ds[0x0040,0xa730][13]
  expsure_time2 = expsure_time1[0x0040,0xa730][5]
  expsure_time3 = expsure_time2[0x0040,0xa730][0]
  expsure_time4 = expsure_time3[0x0040, 0xa300][0]
  expsure_time4
  expsure_time1 = expsure_time4.NumericValue 
except:
  expsure_time1 = "NOT available"

#datetime
try:
  datetime = ds[0x0040,0xa730][8]
  Content_Date_Time = datetime.DateTime
except:
  Content_Date_Time = "NOT available"

#scanning length1
try:
  ct_par5 = ct_par4[1]
  Scanning_length1 = ct_par5[0x0040,0xa300][0].NumericValue
except:
  Scanning_length1 = "NOT available"

# nominal single width1
try:
  ct_par5 = ct_par4[2]
  Nominal_single_Coll_width1 = ct_par5[0x0040,0xa300][0].NumericValue
except:
  Nominal_single_Coll_width1 = "NOT available"


# nominal total width1
try:
  ct_par5 = ct_par4[3]
  Nominal_total_Coll_width1 = ct_par5[0x0040,0xa300][0].NumericValue
except:
  Nominal_total_Coll_width1 = "NOT available"


#kvp1
try:
  ct_par5 = ct_par4[5]
  ct_par6 = ct_par5[0x0040,0xa730]
  ct_par7 = ct_par6[1]
  KVP1 = ct_par7[0x0040,0xa300][0].NumericValue
except:
  KVP1 = "NOT available"


#maximum Xray1
try:
  ct_par5 = ct_par4[5]
  ct_par6 = ct_par5[0x0040,0xa730]
  ct_par7 = ct_par6[2]
  Max_xray_current1 = ct_par7[0x0040,0xa300][0].NumericValue
except:
  Max_xray_current1 = "Not available"


#xray tube current1
try:
  ct_par5 = ct_par4[5]
  ct_par6 = ct_par5[0x0040,0xa730]
  ct_par7 = ct_par6[3]
  Xray_tbe_current1 = ct_par7[0x0040,0xa300][0].NumericValue
except:
  Xray_tbe_current1 = "Not available"

#Total no of irradiation events
try:
  ct_accumalated1 = ds[0x0040,0xa730][11]
  ct_accumalated2 = ct_accumalated1[0x0040,0xa730]
  ct_accumalated3 = ct_accumalated2[0]
  ct_accumalated = ct_accumalated3[0x0040,0xa300][0]
  Total_number_of_Irradiation_Events = ct_accumalated.NumericValue
except:
  Total_number_of_Irradiation_Events = "Not available"
     
#ct dose length product total
try:
  ct_accumalated2 = ct_accumalated1[0x0040,0xa730]
  ct_accumalated4 = ct_accumalated2[1]
  CT_Dose_Length_Product_Total = ct_accumalated4[0x0040,0xa300][0]
  CT_dose_length_total = CT_Dose_Length_Product_Total.NumericValue
except:
  CT_dose_length_total = "Not available"


#ct Acq protocol1
try:
  acq_protocol1 = ds[0x0040,0xa730][12]
  acq_protocol2 = acq_protocol1[0x0040,0xa730]
  acq_protocol3 = acq_protocol2[0]
  Acquisition_Protocol = acq_protocol3.TextValue
except:
  Acquisition_Protocol = "Not available"


#ct Acq type(1)
try:
  acq_protocol1 = ds[0x0040,0xa730][12]
  acq_protocol2 = acq_protocol1[0x0040,0xa730]
  acq_protocol3 = acq_protocol2[2]
  ct_acq_type = acq_protocol3[0x0040,0xa168][0]
  Ct_Acquisition_type1 = ct_acq_type.CodeMeaning +"("+ct_acq_type.CodingSchemeDesignator + ct_acq_type.CodeValue+")"
except:
  Ct_Acquisition_type1 = "Not available"


#ct Acq type(2)
try:
  acq_protocol2 = ct_acq_type2[2]
  ct_type2 = acq_protocol2[0x0040,0xa168][0]
  Ct_Acquisition_type2 = ct_type2.CodeMeaning + '(' + ct_type2.CodeValue + "," +ct_type2.CodingSchemeDesignator + ')'
except:
  Ct_Acquisition_type2 = "Not available"


#CTDIw_Phantom_type1
try:
  ct1 = ds[0x0040,0xa730][12]
  ct2 = ct1[0x0040,0xa730][6]
  ct3 = ct2[0x0040,0xa730][1]
  ct4 = ct3[0x0040,0xa168]
  ct_dose = ct4[0]
  CTDIw_Phantom_type1 = ct_dose.CodeMeaning + '(' + ct_dose.CodeValue + ',' + ct_dose.CodingSchemeDesignator + ')'
except:
  CTDIw_Phantom_type1 = "Not available"


#mean CTDIvoI1
try:
  mean = ct2[0x0040,0xa730][0]
  mean_CTDIvoI = mean[0x0040,0xa300][0]
  Mean_CTDIvoI1 = mean_CTDIvoI.NumericValue
except:
  Mean_CTDIvoI1 = "Not available"

     
#DLP1
try:
  dlp1 = mean = ct2[0x0040,0xa730][2]
  dlp2 = dlp1[0x0040,0xa300][0]
  DLP1 = dlp2.NumericValue
except:
  DLP1 = "Not available"

#ct acq parameter 2
try:
  acq_protocol2 = ds[0x0040,0xa730][13]
  acq_protocol21 = acq_protocol2[0x0040,0xa730]
  acq_protocol22 = acq_protocol21[5]
  acq_protocol23 = acq_protocol22[0x0040,0xa730]
except:
  acq_protocol23 = "Not available"
     
     
#exposure time2
try:
  exp_time2 = acq_protocol23[0]
  exp_time21 = exp_time2[0x0040,0xa300]
  Exposure_time2 = exp_time21[0].NumericValue
except:
  Exposure_time2 ="Not available"
    
#scanning length2 
try:
  scanlen2 = acq_protocol23[1]
  scanlen21 = scanlen2[0x0040,0xa300]
  Scanning_length2 = scanlen21[0].NumericValue
except:
  Scanning_length2 = "Not available"

#exposed range2
try:
  exp_range2 = acq_protocol23[2]
  exp_range22 = exp_range2[0x0040,0xa300]
  Exposed_range = exp_range22[0].NumericValue
except:
  Exposed_range = "Not available"

#nominal single width2
try:
  nominal_single_width2 = acq_protocol23[3]
  nominal_single_width22 =nominal_single_width2[0x0040,0xa300]
  Nominal_single_Coll_width2 = nominal_single_width22[0].NumericValue
except:
  Nominal_single_Coll_width2 = "Not available"

#nominal total width2
try:
  nominal_total_width2 = acq_protocol23[4]
  nominal_total_width22 =nominal_total_width2[0x0040,0xa300]
  Nominal_total_Coll_width2 = nominal_total_width22[0].NumericValue
except:
  Nominal_total_Coll_width2 = "Not available"

#pitch factor2
try:
  pitch_factor2 = acq_protocol23[5]
  pitch_factor22 = pitch_factor2[0x0040,0xa300]
  Pitch_Factor = pitch_factor22[0].NumericValue
except:
  Pitch_Factor = "Not available"

#kvp2
try:
  kvp2 = acq_protocol23[7]
  kvp21 = kvp2[0x0040,0xa730]
  kvp22 = kvp21[1]
  kvp23 = kvp22[0x0040,0xa300]
  KVP2 = kvp23[0].NumericValue
except:
  KVP2 = "Not available"

# maximum Xray tube current2
try:
  xray_tube_curr2 = kvp21[2]
  xray_tube_curr3 = xray_tube_curr2[0x0040,0xa300]
  Max_xray_current2 = xray_tube_curr3[0].NumericValue
except:
  Max_xray_current2 = "Not available"


# Xray tube current2
try:
  xray_tube2 = kvp21[3]
  xray_tube21 = xray_tube2[0x0040,0xa300]
  Xray_tbe_current2 = xray_tube21[0].NumericValue
except:
  Xray_tbe_current2 = "Not available"

#exposure time per rotation
try:
  exp_time_per_rot2 = kvp21[4]
  exp_time_per_rot21 = exp_time_per_rot2[0x0040,0xa300]
  Exposure_Time_per_rotation = exp_time_per_rot21[0].NumericValue
except:
  Exposure_Time_per_rotation = "Not available"
     
#CT Dose 2
try:
  ct_dose21 = ds[0x0040,0xa730][13]
  ct_dose22 = ct_dose21[0x0040,0xa730]
  ct_dose2 = ct_dose22[6]
except:
  ct_dose2 = "Not available"


#mean ctdivoi2
try:
  mean_CTDIvoI21 = ct_dose2
  mean_CTDIvoI22 = mean_CTDIvoI21[0x0040,0xa730]
  mean_CTDIvoI23 = mean_CTDIvoI22[0]
  mean_CTDIvoI24 = mean_CTDIvoI23[0x0040,0xa300]
  Mean_CTDIvoI2 = mean_CTDIvoI24[0].NumericValue
except:
  Mean_CTDIvoI2 = "Not available"


#ctdIw phantom type2
try:
  phantom_type21 = mean_CTDIvoI22[1]
  phantom_type22 = phantom_type21[0x0040,0xa168]
  CTDIw_Phantom_type2 = phantom_type22[0].CodeMeaning + '(' + phantom_type22[0].CodeValue  + ',' +phantom_type22[0].CodingSchemeDesignator+')'
except:
  CTDIw_Phantom_type2 = "Not available"


#Dlp2
try:
  dlp21 = mean_CTDIvoI22[2]
  dlp22 = dlp21[0x0040,0xa300]
  DLP2 = dlp22[0].NumericValue
except:
  DLP2 = "Not available"
```

```python
filecsv = open('DICOM.csv', 'w',encoding='utf8')
csv_columns = ['File name','Study','Protocol','Manufacturar','Content Date/Time','Device Observar physical location',
           'Total number of Irradiation Events','CT dose length total','Acquisition Protocol','Ct Acquisition type1',
      'Exposure time1','Scanning length1','Nominal single Coll width1','Nominal total Coll width1','KVP1','Max xray current1',
      'Xray tbe current1','Mean CTDIvoI1','CTDIw Phantom type1','DLP1','Ct Acquisition type2','Exposure time2','Scanning length2',
      'Exposed range','Nominal single Coll width2','Nominal total Coll width2','Pitch Factor','KVP2','Max xray current2',
      'Xray tbe current2','Exposure Time per rotation','Mean CTDIvoI2','CTDIw Phantom type2','DLP2']
               

writer = csv.DictWriter(filecsv, fieldnames=csv_columns)
writer.writeheader()
```
